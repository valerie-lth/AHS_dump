```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# predicted_results_ensemble.csv have the inference probs for 30K labels
mass_df = pd.read_csv('mass_label_predictions.csv')
met_df = pd.read_csv('met_label_x_predictions.csv')
agg_df = pd.read_csv('aggressive_label_predictions.csv')
```

```python
all_samples = pd.read_csv('sampling_analysis2.csv')
all_mass_dist = pd.DataFrame(all_samples['mass_event_bin'].value_counts().sort_index()).reset_index()
all_mass_dist.columns = ['mass_event_bin', 'Full Sample Size']
all_met_dist = pd.DataFrame(all_samples['met_event_bin'].value_counts().sort_index()).reset_index()
all_met_dist.columns = ['met_event_bin', 'Full Sample Size']
all_agg_dist = pd.DataFrame(all_samples['aggressive_event_bin'].value_counts().sort_index()).reset_index()
all_agg_dist.columns = ['aggressive_event_bin', 'Full Sample Size']
```

```python
mass_df['mass_event_bin'] = pd.cut(mass_df['Probabilities'], bins=[0,0.1,0.4,0.5,0.6,0.9,1])
met_df['met_event_bin'] = pd.cut(met_df['Probabilities'], bins=[0,0.1,0.4,0.5,0.6,0.9,1])
agg_df['aggressive_event_bin'] = pd.cut(agg_df['Probabilities'], bins=[0,0.1,0.4,0.5,0.6,0.9,1])
```
